## Paso 1

Consolidar todos los archivos en un solo csv para analizar un solo dataset

```python
import os
import pandas as pd

# Ruta de la carpeta principal
carpeta_principal = r'C:\Users\Trabajo\Desktop\Sebas\Proyectos\Headphones\AutoEq-master\measurements'

# Lista para almacenar los datos
datos_audifonos = []

# Recorre las carpetas y archivos
for carpeta_evaluador in os.listdir(carpeta_principal):
    carpeta_evaluador_path = os.path.join(carpeta_principal, carpeta_evaluador)
    
    # Verificar si es una carpeta
    if os.path.isdir(carpeta_evaluador_path):
        
        # Añadir "data" a la ruta
        carpeta_data_path = os.path.join(carpeta_evaluador_path, 'data')
        
        # Verificar si existe la carpeta "data"
        if os.path.exists(carpeta_data_path) and os.path.isdir(carpeta_data_path):
            
            # Recorrer las carpetas dentro de "data"
            for tipo_audifono in os.listdir(carpeta_data_path):
                tipo_audifono_path = os.path.join(carpeta_data_path, tipo_audifono)
                
                # Verificar si es una carpeta
                if os.path.isdir(tipo_audifono_path):
                    for archivo_csv in os.listdir(tipo_audifono_path):
                        if archivo_csv.endswith('.csv'):
                            ruta_csv = os.path.join(tipo_audifono_path, archivo_csv)
                            marca_referencia = archivo_csv.split('.')[0]
                            df = pd.read_csv(ruta_csv)
                            frecuencia = df['frequency'].tolist()
                            respuesta = df['raw'].tolist()
                            datos_audifonos.append({'Evaluador': carpeta_evaluador, 'TipoAudifono': tipo_audifono, 'MarcaReferencia': marca_referencia, 'Frecuencia': frecuencia, 'Respuesta': respuesta})

# Crear un DataFrame con los datos
df_final = pd.DataFrame(datos_audifonos)
```
